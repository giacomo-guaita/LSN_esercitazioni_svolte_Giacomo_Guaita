caso liquido


```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

epot = np.loadtxt("epot_inst_liquid.dat",dtype=float)
epot = epot / 108
numlanci = np.arange(epot.shape[0])


plt.plot(numlanci,epot)
plt.xlabel('Throw number')
plt.ylabel('Epot')
plt.grid(True)
plt.show()
```

```python

# Mean
mean = np.mean(epot)

# Variance
var = np.var(epot)

# Normalized data
ndata = epot - mean

acorr = np.correlate(ndata, ndata, 'full')[len(ndata)-1:] 
acorr = acorr / var / len(ndata)


plt.plot(numlanci,acorr)
plt.xlabel('Throw number')
plt.ylabel('Autocorrelations')
plt.grid(True)
plt.xlim(0,1000)
plt.show()
```

```python
L = np.arange(10,5000,10)
N = epot.shape[0]/L
uncertainty = np.zeros(L.shape[0])

for i in range(0,L.shape[0]):
    mean = zeros(N.shape[0])
    for j in range(0,N[i]):
        for z in range(j*L[i],(j+1)*L[i])
            mean[j] += epot[z]/L[i]
    uncertainty[i] = np.std(mean)/sqrt(L[i])

    
plt.plot(L,uncertainty)
plt.xlabel('Points per block')
plt.ylabel('Uncertainty')
plt.grid(True)
plt.show()
    
```
