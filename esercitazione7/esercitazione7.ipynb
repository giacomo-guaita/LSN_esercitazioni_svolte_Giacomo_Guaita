# Esercizio 7.2

Come già accennato nell'esercitazione 4 ho aggiunto l'implementazione dell'algoritmo di Metropolis al codice, si vuole determinare qual è il numero più opportuno di passi per ogni blocco per l'algoritmo. Si vuole evitare l'autocorrelazione tra i blocchi e stuiare l'andamento delle incertezze in funzione del numero di passi.  
Sotto i grafici istantanei dell'energia potenziale per particella per simulazioni equilibrate nei casi solido, liquido e gassoso.

caso liquido


```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

epot_l = np.loadtxt("epot_inst_liquid.dat",dtype=float)
epot_l = epot_l / 108
numlanci_l = np.arange(epot_l.shape[0])

epot_s = np.loadtxt("epot_inst_solid.dat",dtype=float)
epot_s = epot_s / 108
numlanci_s = np.arange(epot_s.shape[0])

epot_g = np.loadtxt("epot_inst_gas.dat",dtype=float)
epot_g = epot_g / 108
numlanci_g = np.arange(epot_g.shape[0])



axes[0,0].plot(numlanci_s,epot_s)
axes[0,0].set_title('Caso solido')
axes[0,0].set_xlabel('$N_{passo}$')
axes[0,0].set_ylabel('$U/N$')
axes[0,0].grid(True)

axes[0,1].plot(numlanci_s,epot_s)
axes[0,1].set_title('Caso liquido')
axes[0,1].set_xlabel('$N_{passo}$')
axes[0,1].set_ylabel('$U/N$')
axes[0,1].grid(True)

axes[1,0].plot(numlanci_g,epot_g)
axes[1,0].set_title('Caso gassoso')
axes[1,0].set_xlabel('$N_{passo}$')
axes[1,0].set_ylabel('$U/N')
axes[1,0].grid(True)

axes[1,1].remove()


plt.tight_layout()
plt.show
```

Calcolo e grafico ora nei tre casi l'autocorrelazione al variare del numero del passo per i blocchi, l'obiettivo è ottenere una autocorrelazione bassa, per far questo saranno necessari blocchi di lunghezza sufficiente.

```python

# Mean
mean_s = np.mean(epot_s)
mean_l = np.mean(epot_l)
mean_g = np.mean(epot_g)

#variance
var_s = np.var(epot_s)
var_l = np.var(epot_l)
var_g = np.var(epot_g)


ndata_s = epot_s - mean_s
ndata_l = epot_l - mean_l
ndata_g = epot_g - mean_g

acorr_s = np.correlate(ndata_s, ndata_s, 'full')[len(ndata_s)-1:] 
acorr_l = np.correlate(ndata_l, ndata_l, 'full')[len(ndata_l)-1:] 
acorr_g = np.correlate(ndata_g, ndata_g, 'full')[len(ndata_g)-1:] 

acorr_s = acorr_s / var_s / len(ndata_s)
acorr_l = acorr_l / var_l / len(ndata_l)
acorr_g = acorr_g / var_g / len(ndata_g)

```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0,0].plot(numlanci_s,acorr_s)
axes[0,0].set_title('Caso solido')
axes[0,0].set_xlabel('$M$ - passi per blocco')
axes[0,0].set_ylabel('autocorrelazione')
axes[0,0].grid(True)
axes[0,0].set_xlim(0,600)

axes[0,1].plot(numlanci_s,acorr_s)
axes[0,1].set_title('Caso liquido')
axes[0,1].set_xlabel('$M$ - passi per blocco')
axes[0,1].set_ylabel('autocorrelazione')
axes[0,1].grid(True)
axes[0,1].set_xlim(0,600)

axes[1,0].plot(numlanci_g,acorr_g)
axes[1,0].set_title('Caso gassoso')
axes[1,0].set_xlabel('$M$ - passi per blocco')
axes[1,0].set_ylabel('autocorrelazione')
axes[1,0].grid(True)
axes[1,0].set_xlim(0,600)

axes[1,1].remove()


plt.tight_layout()
plt.show()
```

Si osserva che per eliminare quasi completamente l'autocorrelazione sono sufficienti blocchi da 100 passi l'uno, sotto questa soglia è però presente autocorrelazione tra i dati, dovuta alla pseudo-randomicità del generatore di numeri casuali.  
Studio adesso l'andamento dell'errore sulla miglior stima dell'energia potenziale per particella in funzione del numero di passi per blocco.

```python
L = np.arange(10,5000,10)
#in caso il programma ci metta troppo ad eseguire si può incrementae il passo del campionamento

#calcolo il numero di blocchi
N_s = epot_s.shape[0]/L
N_l = epot_l.shape[0]/L
N_g = epot_g.shape[0]/L

#inizializzando i vettori delle incertezze
uncertainty_s = np.zeros(L.shape[0])
uncertainty_l = np.zeros(L.shape[0])
uncertainty_g = np.zeros(L.shape[0])

#ciclo su tutte le possibili dimensioni dei blocchi
for i in range(0,L.shape[0]):
    mean = np.zeros(int(N_s[i]))
    #ciclo sul numero di blocchi
    for j in range(0,int(N_s[i])):
        #ciclo sui passi del blocco
        for z in range(j*L[i],(j+1)*L[i]):
            mean[j] += epot_s[z]/L[i]
    uncertainty_s[i] = np.std(mean)/np.sqrt(N_s[i])
    
for i in range(0,L.shape[0]):
    mean = np.zeros(int(N_l[i]))
    for j in range(0,int(N_l[i])):
        for z in range(j*L[i],(j+1)*L[i]):
            mean[j] += epot_l[z]/L[i]
    uncertainty_l[i] = np.std(mean)/np.sqrt(N_l[i])
    
for i in range(0,L.shape[0]):
    mean = np.zeros(int(N_g[i]))
    for j in range(0,int(N_g[i])):
        for z in range(j*L[i],(j+1)*L[i]):
            mean[j] += epot_g[z]/L[i]
    uncertainty_g[i] = np.std(mean)/np.sqrt(N_g[i])

```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0,0].plot(L,uncertainty_s)
axes[0,0].set_title('Caso solido')
axes[0,0].set_xlabel('$M$ - passi per blocco')
axes[0,0].set_ylabel('incertezza')
axes[0,0].grid(True)


axes[0,1].plot(L,uncertainty_l)
axes[0,1].set_title('Caso liquido')
axes[0,1].set_xlabel('$M$ - passi per blocco')
axes[0,1].set_ylabel('incertezza')
axes[0,1].grid(True)


axes[1,0].plot(L,uncertainty_g)
axes[1,0].set_title('Caso gassoso')
axes[1,0].set_xlabel('$M$ - passi per blocco')
axes[1,0].set_ylabel('incertezza')
axes[1,0].grid(True)


axes[1,1].remove()


plt.tight_layout()
plt.show()
```

Si nota che dopo un certo numero di passi per blocco, che dipende dal fase della mateia studiata, l'incertezza si assesta anche se con importanti fluttuazioni attorno ad un valore costante, che è indipendente dal numero di passi per blocco. Questa è un'indicazione che la media e l'incertezza calcolate a blocchi sono indipendenti dalla dimensione del blocco ma dipendono solo dal numero totale di passi e dal problema studiato, sono dunque un valido strumento per le simulazioni numeriche. Il valore dell'incertezza per piccoli numeri di passi è molto basso a causa dell'autocorrelazione tra questi blocchi, si nota però che il numero di passi dopo il quale l'incertezza converge è maggiore rispetto al numero di passi dopo il quale l'autocorrelazione converge.
