3. Ho estratto 10000 numeri pseudo-casuali nell'intervallo $[0,1)$ e li ho suddivisi in 100 bin di egual dimensione, per ognuno di questi bin ho calcolato il chi2 con la formula di Pearson.
$$\chi^2 = \sum_{i} \frac{\left( O_i - E_i \right)^2}{E_i}$$
Dove $O_i$ è il numero di estrazioni pseudo-casuali nell'i-esimo bin, mentre $E_i$ è il valore di aspettazione di queste estrazioni nell'i-esimo bin.  
Sotto ho fatto l'istogramma dei chi2 ottenuti in questo modo nei vari bin. Si nota che esso è centrato attorno al valore 100, che è quanto ci aspettiamo per una distribuzione casuale uniforme. Ho inoltre stampato il valor medio e la deviazione standard di questi chi2 per sottolineare ciò.

```python
chi2 = np.loadtxt("Chi2.dat",dtype=float)
plt.hist(chi2, color='orange')
plt.grid(axis='y')
plt.xlabel('$\chi^2$')
plt.ylabel('Frequenza osservata')
plt.title('Istogramma del $\chi^2$ dei bin')
plt.show()

print('Media chi2 = ',np.mean(chi2))
print('Deviazione standard chi2 = ', np.std(chi2))
```

# Esercizio 1.2

Ho aggiunto due metodi alla classe random per estrarre numeri generati in una distribuzione esponenziale o lorentziana, essi vengono estratti con il metodo della funzione cumulativa.  
Con questi metodi ho prodotto istogrammi con $10^4$ realizzazioni di $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ per N=1,2,10,100 delle seguenti tre distribuzioni di probabilità: esponenziale, dado a sei facce e lorentziana.

1. Distribuzione esponenziale:

```python
exp1 = np.loadtxt("Exp.dat",skiprows=1,max_rows=10000,dtype=float)
exp2 = np.loadtxt("Exp.dat",skiprows=10005,max_rows=10000,dtype=float)
exp3 = np.loadtxt("Exp.dat",skiprows=20009,max_rows=10000,dtype=float)
exp4 = np.loadtxt("Exp.dat",skiprows=30013,dtype=float)
(mu,sigma)=norm.fit(exp4)



fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].hist(exp1, color='blue', alpha=0.7)
axes[0, 0].set_title('N = 1')
axes[0, 0].set_xlabel('$S_N$')

axes[0, 1].hist(exp2, color='green', alpha=0.7)
axes[0, 1].set_title('N = 2')
axes[0, 1].set_xlabel('$S_N$')

axes[1, 0].hist(exp3, color='red', alpha=0.7)
axes[1, 0].set_title('N = 10')
axes[1, 0].set_xlabel('$S_N$')

n, bins, patches = axes[1, 1].hist(exp4, color='purple', bins=30, alpha=0.7, density=True)
axes[1, 1].set_title('N = 100')
axes[1, 1].set_xlabel('$S_N$')

y = norm.pdf( bins, mu, sigma)
axes[1, 1].plot(bins, y, 'r--', linewidth=2)

plt.tight_layout()
fig.suptitle('Istogrammi delle medie di N estrazioni da una distribuzione esponenziale', y=1.05)
fig.subplots_adjust(top=0.92)
plt.show()
```

2. Dado:

```python
Dice1 = np.loadtxt("Dice.dat",skiprows=1,max_rows=10000,dtype=float)
Dice2 = np.loadtxt("Dice.dat",skiprows=10005,max_rows=10000,dtype=float)
Dice3 = np.loadtxt("Dice.dat",skiprows=20009,max_rows=10000,dtype=float)
Dice4 = np.loadtxt("Dice.dat",skiprows=30013,dtype=float)
(mu,sigma)=norm.fit(Dice4)



fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].hist(Dice1, bins=6, color='blue',edgecolor='black', alpha=0.7)
axes[0, 0].set_title('N = 1')
axes[0, 0].set_xlabel('$S_N$')

axes[0, 1].hist(Dice2, color='green', alpha=0.7)
axes[0, 1].set_title('N = 2')
axes[0, 1].set_xlabel('$S_N$')

axes[1, 0].hist(Dice3, bins=15, color='red', alpha=0.7)
axes[1, 0].set_title('N = 10')
axes[1, 0].set_xlabel('$S_N$')

n, bins, patches = axes[1, 1].hist(Dice4, color='purple', bins=30, alpha=0.7, density=True)
axes[1, 1].set_title('N = 100')
axes[1, 1].set_xlabel('$S_N$')

y = norm.pdf( bins, mu, sigma)
axes[1, 1].plot(bins, y, 'r--', linewidth=2)

plt.tight_layout()
fig.suptitle('Istogrammi delle medie di N lanci di dado', y=1.05)
fig.subplots_adjust(top=0.92)
plt.show()
```

Per questi primi due casi si nota che grazie al teorema del limite centrale all'aumentare di N, la distribuzione tende a quella gaussiana. Nel caso N=100 ho sovrapposto all'istogramma una linea rossa tratteggiata con la gaussiana che fitta meglio l'istogramma. La validità di questi fit è un esempio dell'estrema utilità del teorema del limite centrale.

3. Distribuzione lorentziana

```python
Lorentz1 = np.loadtxt("Lorentz.dat",skiprows=1,max_rows=10000,dtype=float)
Lorentz2 = np.loadtxt("Lorentz.dat",skiprows=10005,max_rows=10000,dtype=float)
Lorentz3 = np.loadtxt("Lorentz.dat",skiprows=20009,max_rows=10000,dtype=float)
Lorentz4 = np.loadtxt("Lorentz.dat",skiprows=30013,dtype=float)

bins=range(-10,11,1)
thinbins=np.arange(-10,11,0.5)

median=np.quantile(Lorentz4,0.5)
scale=(np.quantile(Lorentz4,0.85)-np.quantile(Lorentz4,0.15))/2

cauchy=scale/(np.pi*(thinbins-median)**2+scale**2)*0.65



fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].hist(Lorentz1, bins, color='blue', alpha=0.7)
axes[0, 0].set_title('N = 1')
axes[0, 0].set_xlabel('$S_N$')

axes[0, 1].hist(Lorentz2, bins, color='green', alpha=0.7)
axes[0, 1].set_title('N = 2')
axes[0, 1].set_xlabel('$S_N$')

axes[1, 0].hist(Lorentz3, bins, color='red', alpha=0.7)
axes[1, 0].set_title('N = 10')
axes[1, 0].set_xlabel('$S_N$')

n, bins, patches = axes[1, 1].hist(Lorentz4, thinbins, color='purple', alpha=0.7, density=True)
axes[1, 1].set_title('N = 100')
axes[1, 1].set_xlabel('$S_N$')
axes[1, 1].plot(thinbins,cauchy, 'r--', linewidth=2)


plt.tight_layout()
fig.suptitle('Istogrammi delle medie di N lanci di dado', y=1.05)
fig.subplots_adjust(top=0.92)
plt.show()
```

Per il caso della lorentziana si nota che la forma dell'istogramma rimane invariata a prescindere dal numero di di distribuzioni che si sommano insieme. Infatti fittando il caso N=100 con una lorentziana (tratteggiata in rosso) si nota che è un buon fit. Risulta anche evidente che la somma di lorentziane non possa mai convergere a una gaussiana. Infatti la distribuzione lorentziana non soddisfa l'ipotesi del teorema del limite centrale, nella sua forma più semplice, di avere varianza e quindi integrale su $\mathbb{R}$ finito. Esistono estensioni del teorema del limite centrale per cui somme di distribuzioni con caratteristiche specifiche convergono ad alcune distribuzioni peculiari, si noti che la lorentziana è una di queste ed è la funzione di convergenza di se stessa.

# Esercizio 1.3

Ho calcolato una stima di pigreco usando il metodo di Buffon, ho simulato il lancio di una sbarretta di lunghezza $L$ in un campo di linee parallele a distanza $d$ tra di loro, con $d>L$. La probabilità che la sbarretta tocchi una linea in questo lancio è data dalla seguente formula.
$$ P = \frac{2 L}{\pi d} $$
Invertendo la formula si può ricavare il valore di pigreco. Ho simulato il lancio della sbarretta senza l'uso di pigreco per blocchi da $10^3$ lanci l'uno, e per ogni blocco ho calcolato la probabilità che la sbarretta colpisca la linea come $P=N_{hit}/N$.  
Sotto il grafico della mia miglior stima di pigreco con la relativa incertezza. In rosso tratteggiato il valore di pigreco preso da numpy.

```python
media = np.loadtxt("Pi.dat",usecols=1,skiprows=1,dtype=float)
incertezza = np.loadtxt("Pi.dat",usecols=2,skiprows=1,dtype=float)
numlanci = np.loadtxt("Pi.dat",usecols=0,skiprows=1,dtype=int)
L = np.loadtxt("Pi.dat",dtype=int,max_rows=1)



plt.errorbar(numlanci,media,yerr=incertezza)
plt.axhline(y=np.pi, color='r', linestyle='--', linewidth=3)
plt.xlabel('#blocchi')
plt.ylabel('stima $\pi$')
plt.grid(True)
plt.xlim(0,500)
plt.show()


print('Dimensione blocchi: ',L)
```

Con questo metodo si può ottenere una buona stima di pigreco anche se la convergenza non è molto rapida.
