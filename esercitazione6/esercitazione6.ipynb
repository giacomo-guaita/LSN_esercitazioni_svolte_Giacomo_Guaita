# Esercizio 6.1

Ho completato il codice con la misura di capacità termica, sucettibilità magnetica e magnetizzazione, ho implementato l'algoritmo sia con Metropolis sia con il campinamento alla Gibbs. Inoltre ho aggiunto un file di output data_out nel quale stampo: temperatura, energia, capacità termica, suscettività magnetica e magnetizzazione e le relative incertezze; questo file di output è utile per graficare le quantità richieste e utilizza medie globali su tutta la simulazione.  
Ho quindi eseguito simulazioni a temperature successive on finezza di campionamento $\delta T=0.1$ per le temperature nell'intervallo $T=[0.5,2]$, raccogliendo dati per graficare le quantità richieste in funzione della temperatura in questo intevallo con un grosso numero di passi $n_{steps}=2 \cdot 10^5$.  
Prima di simulare uno di questi blocchi ho equilibrato la simulazione: questo è necessario perché la configurazione casuale di spin, da cui il programma parte di default, non è rappresentativa soprattutto a basse temperature per basse temperature. L'equilibrazione è stata svolta in questo modo: ho eseguito il programma da una configurazione casuale per un numero limitato di blocchi(ne ho usati 5 da $10^4$ passi l'uno) alla temperatura di $T=0.5$, e ho poi eseguito in successione le simulazioni con più blocchi per temperature successive nell'intervallo richiesto, utilizzando ogni volta come configurazione iniziale quella finale della simulazione precedente. Questa è un'approssimazione perché a parte la temperatura iniziale $T=0.5$ tutte le altre partono da configurazioni tipiche di temperature diverse, ma siccome utilizzo temperature vicine il transiente di equilibrazione è molto corto e non dovrebbe impattare sulle misure complessive della simulazione.

Sotto i grafici delle quantità richieste con le relative incertezze, simulati con Metropolis e con Gibbs e raffrontati ai valori attesi dal modello teorico. 

1. Energia interna h=0

```python
energiaM = np.loadtxt("h=0_metro.dat",usecols=1,dtype=float)
incertezzaM = np.loadtxt("h=0_metro.dat",usecols=2,dtype=float)
temp = np.loadtxt("h=0_metro.dat",usecols=0,dtype=float)

energiaG = np.loadtxt("h=0_gibbs.dat",usecols=1,dtype=float)
incertezzaG = np.loadtxt("h=0_gibbs.dat",usecols=2,dtype=float)



beta = 1/temp
J = 1.0
Ns = 50
th = np.tanh(J/temp)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

fig, axes = plt.subplots(2, 1, figsize=(10, 8))

axes[0].plot(temp, e, '--r',linewidth=2.5,label='Modello teorico')
axes[0].errorbar(temp,energiaM,yerr=incertezzaM,label='Metropolis')
axes[0].errorbar(temp,energiaG,yerr=incertezzaM, label='Gibbs')
axes[0].set_xlabel('Temperatura')
axes[0].set_ylabel('Energia')
axes[0].set_title('Ising 1D, energia interna')
axes[0].legend()
axes[0].grid(True)

axes[1].errorbar(temp,energiaM-e,yerr=incertezzaM,label='Metropolis')
axes[1].errorbar(temp,energiaG-e,yerr=incertezzaM,label='Gibbs')
axes[1].set_xlabel('Temperatura')
axes[1].set_ylabel('Energia')
axes[1].set_title('Differenza di energia tra le simulazioni e il modello')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

Si osserva che a basse energie l'algoritmo di metropolis è più accurato del campionamento di Gibbs, ma entrambe le simulazioi sono in accordo con il modello teorico. Si osserva anche che le incertezze simulate diminuiscono lievemente con l'aumentare della temperatura.

2. Capacità termica h=0

```python
HCapacityM = np.loadtxt("h=0_metro.dat",usecols=3,dtype=float)
incertezzaM = np.loadtxt("h=0_metro.dat",usecols=4,dtype=float)
HCapacityG = np.loadtxt("h=0_gibbs.dat",usecols=3,dtype=float)
incertezzaG = np.loadtxt("h=0_gibbs.dat",usecols=4,dtype=float)

heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)



fig, axes = plt.subplots(2, 1, figsize=(10, 8))

axes[0].plot(temp, heat,  '--r',linewidth=2.5,label='Modello teorico')
axes[0].errorbar(temp,HCapacityM,yerr=incertezzaM,label='Metropolis')
axes[0].errorbar(temp,HCapacityG,yerr=incertezzaM, label='Gibbs')
axes[0].set_xlabel('Temperatura')
axes[0].set_ylabel('C')
axes[0].set_title('Ising 1D, capacità termica')
axes[0].legend()
axes[0].grid(True)

axes[1].errorbar(temp,HCapacityM-heat,yerr=incertezzaM,label='Metropolis')
axes[1].errorbar(temp,HCapacityG-heat,yerr=incertezzaM,label='Gibbs')
axes[1].set_xlabel('Temperatura')
axes[1].set_ylabel('C')
axes[1].set_title('Differenza di capacità termica tra le simulazioni e il modello')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

I due modelli ricalcano la forma della curva della capacità termica prevista dal modello, ma risultano essere totalmente incompatibili con esso, poichè distano da esso svariate deviazioni standard. La motivazione può essere ricercata in un possibile errore nel codice. 

3. Suscettibilità magnetica

```python
ChiM = np.loadtxt("h=0_metro.dat",usecols=7,dtype=float)
incertezzaM = np.loadtxt("h=0_metro.dat",usecols=8,dtype=float)
ChiG = np.loadtxt("h=0_gibbs.dat",usecols=7,dtype=float)
incertezzaG = np.loadtxt("h=0_gibbs.dat",usecols=8,dtype=float)


X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)


fig, axes = plt.subplots(2, 1, figsize=(10, 8))

axes[0].plot(temp, X,  '--r',linewidth=2.5,label='Modello teorico')
axes[0].errorbar(temp,ChiM,yerr=incertezzaM,label='Metropolis')
axes[0].errorbar(temp,ChiG,yerr=incertezzaM, label='Gibbs')
axes[0].set_xlabel('Temperatura')
axes[0].set_ylabel('$\chi$')
axes[0].set_title('Ising 1D, suscettività magnetica')
axes[0].legend()
axes[0].grid(True)

axes[1].errorbar(temp,ChiM-X,yerr=incertezzaM,label='Metropolis')
axes[1].errorbar(temp,ChiG-X,yerr=incertezzaM,label='Gibbs')
axes[1].set_xlabel('Temperatura')
axes[1].set_ylabel('$\chi$')
axes[1].set_title('Differenza di suscettività tra le simulazioni e il modello')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

Come nel caso dell'energia interna la simulazione metropolis risulta più accurata a basa temperatura rispetto al campionamento di Gibbs, entrambi i modelli forniscono però una buona previsione del comportamento della suscettività magnetica. Si nota inoltre che entrambi i modelli hanno un'incertezza maggiore a basse temperature, questo fa pensre a una possibile difficoltà dei modelli montecarlo nel descrivere sistemi con basse probabilità di transizioni di stato come quelli a bassa temperatura. In questo caso la diminuzione dell'incertezza assoluta è anche legato alla diminuzione del valore della suscettività, quindi con un errore relativo costante si assiste a una diminuzione dell'errore assoluto, ma la diminuzione delle incertezze si osserva anche indipendentemente da ciò.

4. Magnetizzazione h=0.02

```python
M_simM = np.loadtxt("h=002_metro.dat",usecols=5,dtype=float)
incertezzaM = np.loadtxt("h=002_metro.dat",usecols=6,dtype=float)
M_simG = np.loadtxt("h=002_gibbs.dat",usecols=5,dtype=float)
incertezzaG = np.loadtxt("h=002_gibbs.dat",usecols=6,dtype=float)


h=0.02 #external field
b = 1/temp

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M_es = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)


fig, axes = plt.subplots(2, 1, figsize=(10, 8))

axes[0].plot(temp, M_es,  '--r',linewidth=2.5,label='Modello teorico')
axes[0].errorbar(temp,M_simM,yerr=incertezzaM,label='Metropolis')
axes[0].errorbar(temp,M_simG,yerr=incertezzaM, label='Gibbs')
axes[0].set_xlabel('Temperatura')
axes[0].set_ylabel('M')
axes[0].set_title('Ising 1D, magnetizzazione')
axes[0].legend()
axes[0].grid(True)

axes[1].errorbar(temp,M_simM-M_es,yerr=incertezzaM,label='Metropolis')
axes[1].errorbar(temp,M_simG-M_es,yerr=incertezzaM,label='Gibbs')
axes[1].set_xlabel('Temperatura')
axes[1].set_ylabel('M')
axes[1].set_title('Differenza di magnetizzazione tra le simulazioni e il modello')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

Si osserva un accordo accettabile tra le due simulazioni e il modello teorico, anche in questo caso c'è una diminuzione delle incertezze all'aumentare della temperatura.
