# Esercizio 2.1

Ho eseguito numericamente il seguente integrale, sia con un campionamento uniforme dei punti che tramite importance sampling, ossia utilizzando per le estrazioni una distribuzione di probabilit√† che assomigli maggiormente all'integranda di una distribuzione uniforme.
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

1. Campionamento uniforme  
L'integrale si svolge semplicemente come la media dei valori dell'integranda valutata in punti estratti casualmente, questa media deve poi essere moltiplicata per la dimensione dell'intervallo di integrazione.  
Sotto un grafico dell'andamento dell'integrale calcolato numericamente meno il relativo valore di aspettazione, al progredire del nuemero di blocchi, ovviamente con la relativa incertezza. Ho usato blocchi da 1000 punti ciascuno.

```python
media = np.loadtxt("unif.dat",usecols=1,skiprows=1,dtype=float)
incertezza = np.loadtxt("unif.dat",usecols=2,skiprows=1,dtype=float)
numlanci = np.loadtxt("unif.dat",usecols=0,skiprows=1,dtype=int)
L = np.loadtxt("unif.dat",dtype=int,max_rows=1)


plt.errorbar(numlanci,media-1,yerr=incertezza)
plt.xlabel('#blocchi')
plt.ylabel('$I-1$')
plt.grid(True)
plt.show()

print('Dimensione blocco: ',L)
```

2. Importance sampling
L'importance sampling si basa sull'equivalenza matematica tra i seguenti integrali:
$$ \int_{a}^{b} f(x)dx = \int_{a}^{b} \frac{f(x)}{p(x)} p(x) dx$$
Apparenemente sembra banale e inutile ma se $p(x)$ √® una distribuzione di probabilit√†, il secondo integrale equivale alla media della funzione $g(x)=f(x)/p(x)$ valutata in punti estratti dalla distibuzione p(x) moltiplicata per la dimensione dell'intervallo di integrazione. E grazie a considerazioni matematiche leggermente pi√π complesse si ottiene che l'integrale calcolato √® tanto pi√π preciso quanto pi√π $p(x)$ "assomiglia" a f(x).
Nel caso specifico dell'integrale del coseno ho usato uno sviluppo di Taylor al primo ordine centrato in $x=1$, questo ci permette di calcolare pi√π facilmente la normalizzazione rispetto a uno sviluppo di Taylor al secondo ordine in $x=0$. Inoltre ci permette di invertire questa funzione per campionarne la probabilit√† con il metodo della cumulativa. In questo modo si ottiene $p(x)=-2(x-1)$.  
Sotto un grafico dell'andamento dell'integrale calcolato con importance sampling meno il relativo valore di aspettazione, al progredire del nuemero di blocchi, ovviamente con la relativa incertezza. Ho usato blocchi da 1000 punti ciascuno.

```python
media = np.loadtxt("impsamp.dat",usecols=1,skiprows=1,dtype=float)
incertezza = np.loadtxt("impsamp.dat",usecols=2,skiprows=1,dtype=float)
numlanci = np.loadtxt("impsamp.dat",usecols=0,skiprows=1,dtype=int)
L = np.loadtxt("impsamp.dat",dtype=int,max_rows=1)


plt.errorbar(numlanci,media-1,yerr=incertezza)
plt.xlabel('#blocchi')
plt.ylabel('$I-1$')
plt.grid(True)
plt.show()

print('Dimensione blocco: ',L)
```

Si nota che l'integrale calcolato in questo modo √® pi√π accurato e ha incertezza calcolata a blocchi pi√π piccola. 
Ossia l'integrale calcolato con importance sampling converge pi√π rapidamente.

# Esercizio 2.2

Ho simulato dei cammini casuali in tre dimensioni, su un reticolo cubico di passo $a=1$ con mosse in direzione casuale lungo uno degli assi cartesiani e nel continuo con mosse in direzioni casuali lungo una direzione identificata da due angoli $\theta$ e $\phi$. In entrambi i casi mi aspetto che la distanza media dal punto di partenza abbia un andamento proporzionale a $\sqrt{N}$, dove $N$ √® il numero di passi. Questo √® infatti il tipico andamento dei moti browniani di cui questi random walk sono degli esempi.

1. Cammino su griglia discreta  
Ecco un'immagine di un cammino di questo tipo.

```python
x = np.loadtxt("Draw_xyz.dat",usecols=1,dtype=float)
y = np.loadtxt("Draw_xyz.dat",usecols=2,dtype=float)
z = np.loadtxt("Draw_xyz.dat",usecols=3,dtype=float)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, 'o-', color='green')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

Ecco un'immagine della distanza media dal puno di partenza per $10^4$ simulazione divise in 100 blocchi, la distanza √® graficata rispetto al numero del passo a cui √® calcolata la distanza media, ovviamente la distanza √® graficata con la relativa incertezza. C'√® inoltre un fit con una funzione $f(x)=y_0+k\sqrt{x}$, √® facile osservare che √® un ottimo fit.

```python
def squareroot(x,k,y0):
    y=y0+k*(x)**0.5
    return y

media = np.loadtxt("RWxyz.dat",usecols=1,skiprows=1,dtype=float)
incertezza = np.loadtxt("RWxyz.dat",usecols=2,skiprows=1,dtype=float)
numlanci = np.loadtxt("RWxyz.dat",usecols=0,skiprows=1,dtype=int)
L = np.loadtxt("RWxyz.dat",dtype=int,max_rows=1)



parameters, cov = curve_fit(squareroot, numlanci, media-1)
fit_k = parameters[0]
fit_y0 = parameters[1]


plt.errorbar(numlanci,media-1,yerr=incertezza)
plt.plot(numlanci,squareroot(numlanci,fit_k,fit_y0),'r--',linewidth=4)
plt.xlabel('Numero di passi')
plt.ylabel('<r> xyz')
plt.grid(True)
plt.show()

print('valori del fit per k e y_0 ',fit_k,fit_y0)
print('Dimensione blocco: ',L)
```
