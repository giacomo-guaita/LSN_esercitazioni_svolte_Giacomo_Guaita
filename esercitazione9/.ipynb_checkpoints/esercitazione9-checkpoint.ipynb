# Esercizio 9.1

Ho costruito un algoritmo genetico per la soluzione del problema del messo viaggiatore per città posizionate su una circonferenza oppure dentro ad un quadrato. L'algoritmo si basa sulla costruzione di una popolazione, i cui membri detti cromosomi detengono un possibile cammino che congiunge tutte le città. Questa popolazione viene fatta evolvere attraverso meccanismi di "scambio genetico" tra i cromosomi e mutazioni casuali di singoli cromosomi, è probabilisticamente favorita la "riproduzione" di cromosomi che hanno una lunghezza del cammino tra le città minore. Questo causa una progressiva diminuzione del cammino totale, fino ad arrivare a trovare il miglior percorso possibile tra le città.  
Entro ora più nel dettaglio dell'algoritmo: l'algoritmo non è elitario, questo è stato scelto per permettere una variabilità maggiore dei geni anche tra i migliori cromosomi e la scelta dei cromosomi che si riproducono è comunque sufficientemente top heavy da permettere una convergenza rapida.  Ho costruito un metodo di controllo del percorso tra le città per assicurarmi che nessun cromosoma passasse mai due volte dalla stesssa città o che passasse da città inesistenti. Per fare questo considero il vettore che rappresenta il percorso di un cromosoma, ossia un vettore di numeri interi che rappresentano le città e il loro ordine di apparizione l'ordine in cui esse vengono visitate, e ne sommo tutti gli elementi, questa somma dovrebbe assumere il valore:
$$\sum_{n_{city}=0}^{N}n_{city}=\frac{1}{2}N(N+1)$$
Se questa condizione non è soddisfatta oppure la prima città a comparire nel vettore non è la città numerata 0(ossia numerata come prima città,questo serve a evitare multiple scritture dello stesso percorso), il metodo comunica l'errore in output.  
Ho implementato le seguenti mutazioni:
* scambio di coppie di città
* traslazione di m città accostate di n posizioni
* rotazioni cicliche tra gruppi di città
* inversione dell'ordine tra gruppi di città  

E la seguente tecnica di "ricombinazione genetica": estrarre un numero casuale n e prendere le prime n città di un cromosoma padre e completare il cromosoma figlio con le città rimanenti nell'ordine in cui appaiono nel secondo cromosoma padre. Questa tecnica può produrre due cromosomi figli scambiando i ruoli dei cromosomi padri e mantenendo il numeo estratto n.

Ho testato il mio algoritmo genetico per i due casi di città sulla circonferenza e nel quadrato in un caso con 34 città, l'algoritmo è costruito per usare una popolazione pari al quadrato del numero di città che rimane un numero ragionevole di cromosomi al variare del numero di città. Nei due casi mostrerò grafici della lunghezza media del cammino per la metà migliore della popolazione e per il miglior individuo della popolazione al progredire delle generazioni, e un'immagine del miglior cammino trovato dall'algoritmo nella generazione finale.

1. Città su una circonferenza

```python
mean = np.loadtxt("mean_circ.dat",usecols=1,skiprows=1,dtype=float)
numlanci = np.loadtxt("mean_circ.dat",usecols=0,skiprows=1,dtype=int)
best = np.loadtxt("best_circ.dat",usecols=1,skiprows=1,dtype=float)



plt.plot(numlanci,mean,label='media della metà migliore')
plt.plot(numlanci,best,label='miglior cromosoma')
plt.xlabel('generazione')
plt.ylabel('Lunghezza cammino')
plt.title('Lunghezza L(1) del cammino in funzione della generazione')
plt.legend()
plt.grid(True)
plt.show()
```

Si nota che l'algoritmo converge veocemente, dopo circa  100 generazioni è già arrivato a convergenza, e si nota che l'andamento del miglior cromosoma e l'andamento della miglior metà della popolazione sono molto simili, questo vuol dire che la popolazione tende a "imitare" il miglior cromosoma migliornadosi.

```python
x = np.loadtxt("draw_circ.dat",usecols=0,dtype=float)
y = np.loadtxt("draw_circ.dat",usecols=1,dtype=float)

plt.plot(x,y,'-o',color='purple')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Miglior cammino')
plt.grid(True)
plt.show()
```

L'algoritmo genetico è riuscito a trovare il miglior percorso, che è ovviamente quello che segue la circonferenza.

2. Città dentro un quadrato

```python
mean = np.loadtxt("mean_squ.dat",usecols=1,skiprows=1,dtype=float)
numlanci = np.loadtxt("mean_squ.dat",usecols=0,skiprows=1,dtype=int)
best = np.loadtxt("best_squ.dat",usecols=1,skiprows=1,dtype=float)



plt.plot(numlanci,mean,label='media della metà migliore')
plt.plot(numlanci,best,label='miglior cromosoma')
plt.xlabel('generazione')
plt.ylabel('Lunghezza cammino')
plt.title('Lunghezza L(1) del cammino in funzione della generazione')
plt.legend()
plt.grid(True)
plt.show()
```

Si nota innanzi tutto che il tempo di convergenza è leggermente maggiore rispetto a quello del caso delle città sulla circonferenza, questo è perché il problema è lievemente più complesso. Inoltre si nota in questo caso che ci sono dei piccoli plateau nella lunghezza del cromosoma migliore e delle oscillazioni nella lunghezza media della miglior metà della popolazione prima di arrivare alla convergenza finale. Questi plateau si trovano in corrispondenza di minimi locali della lunghezza del percorso, che gli algoritmi genetici sono un buono strumento per tentare di superare peché studiano una parte molto grande dello spazio dei arametri del problema.

```python
x = np.loadtxt("draw_squ.dat",usecols=0,dtype=float)
y = np.loadtxt("draw_squ.dat",usecols=1,dtype=float)

plt.plot(x,y,'-o',color='purple')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Miglior cammino')
plt.grid(True)
plt.show()
```

Il percorso non si incrocia mai che è una prima indicazione del fatto che sia ottimizzato. La reale certezza che questo sia il percorso ottimale si può avere solo studiando tutti i percorsi possibili, ma la pesantezza computazionale(andamento del tipo $N!$, dove N è il numero di città) di quest' esplorazione è il motivo per cui si decide di affrontare il problema con un algoritmo genetico.
