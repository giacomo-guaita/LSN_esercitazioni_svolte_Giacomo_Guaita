# Esercizi 4.1 e 4.2

Ho completato il codice fornito con l'aggiunta della misura e della stampa della pressione e con la parte di algoritmo di metropolis.  
Ho eseguito il codice per le condizioni richieste:  
1. solido: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquido: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gassoso: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)  


E ho salvato i dati di output in file nella cartella madre con nomi significativi sul loro contenuto. Prima di svolgere le varie simulazioni ho equilibrato il sistema in modo che la temperatura convergesse al valore desiderato, ho fatto questo eseguendo cinque blocchi di simulazione partendo da una configurazione casuale con una temperatura di partenza diversa da quella desiderata e poi eseguendone venti, da cui ho estratto i valori, partendo dalla configurazione ottenuta dalla simulazione dei primi cinque blocchi e con la temperatura di partenza desiderata.

1. caso solido  
Sotto la curva di equilibrazione della temperatura, si nota che essa tende a scendere in maniera importante rispetto alla temperatura iniziale e poi a stabilizzarsi sulla temperatura ridotta desiderata $T*=0.8$. Per questo è necessaria l'equilibrazione del sistema.

```python
equilibration = np.loadtxt("equilibration_solid.dat",dtype=float)
nsteps = range(equilibration.size)

plt.plot(nsteps,equilibration)
plt.xlabel('Steps')
plt.ylabel('Temperature')
plt.title('Curva di equilibrazione della temperatura del solido')
plt.grid(True)
plt.xlim(0,3000)
plt.show()
```

Sotto i grafici degli andamenti di temperatura, energia potenziale, energia cinetica, energia totale e pressione  in unità ridotte di Lennard-Jones con le relative incertezze graficate rispetto al numero del blocco. Sto usando blocchi da 2000 passi l'uno

```python
fig, axes = plt.subplots(3, 2, figsize=(10, 8))

media = np.loadtxt("output_temp_solid.dat",usecols=2,max_rows=40,dtype=float)
incertezza = np.loadtxt("output_temp_solid.dat",usecols=3,max_rows=40,dtype=float)
numlanci = np.loadtxt("output_temp_solid.dat",usecols=0,max_rows=40,dtype=int)

axes[0, 0].errorbar(numlanci,media,yerr=incertezza)
axes[0, 0].set_title('Temperatura')
axes[0, 0].set_xlabel('$N blocco$')
axes[0, 0].set_ylabel('$T*$')
axes[0, 0].grid(True)

media = np.loadtxt("output_ekin_solid.dat",usecols=2,max_rows=40,dtype=float)
incertezza = np.loadtxt("output_ekin_solid.dat",usecols=3,max_rows=40,dtype=float)
numlanci = np.loadtxt("output_ekin_solid.dat",usecols=0,max_rows=40,dtype=int)

axes[0, 1].errorbar(numlanci,media,yerr=incertezza)
axes[0, 1].set_title('Energia cinetica')
axes[0, 1].set_xlabel('$N blocco$')
axes[0, 1].set_ylabel('$E_{kin}*$')
axes[0, 1].grid(True)

media = np.loadtxt("output_epot_solid.dat",usecols=2,max_rows=40,dtype=float)
incertezza = np.loadtxt("output_epot_solid.dat",usecols=3,max_rows=40,dtype=float)
numlanci = np.loadtxt("output_epot_solid.dat",usecols=0,max_rows=40,dtype=int)

axes[1, 0].errorbar(numlanci,media,yerr=incertezza)
axes[1, 0].set_title('Energia potenziale')
axes[1, 0].set_xlabel('$N blocco$')
axes[1, 0].set_ylabel('$E_{pot}*$')
axes[1, 0].grid(True)

media = np.loadtxt("output_etot_solid.dat",usecols=2,max_rows=40,dtype=float)
incertezza = np.loadtxt("output_etot_solid.dat",usecols=3,max_rows=40,dtype=float)
numlanci = np.loadtxt("output_etot_solid.dat",usecols=0,max_rows=40,dtype=int)

axes[1, 1].errorbar(numlanci,media,yerr=incertezza)
axes[1, 1].set_title('Energia totale')
axes[1, 1].set_xlabel('$N blocco$')
axes[1, 1].set_ylabel('$E_{tot}*$')
axes[1, 1].grid(True)

media = np.loadtxt("output_press_solid.dat",usecols=2,max_rows=40,dtype=float)
incertezza = np.loadtxt("output_press_solid.dat",usecols=3,max_rows=40,dtype=float)
numlanci = np.loadtxt("output_press_solid.dat",usecols=0,max_rows=40,dtype=int)

axes[2, 0].errorbar(numlanci,media,yerr=incertezza)
axes[2, 0].set_title('Pressione')
axes[2, 0].set_xlabel('$N blocco$')
axes[2, 0].set_ylabel('$P*$')
axes[2, 0].grid(True)

axes[2, 1].remove()

plt.tight_layout()
plt.show()
```

3. caso gassoso  
Sotto la curva di equilibrazione della temperatura, si nota che rispetto all'equilibrazione di liquido e solido quella del gas è meno immediata e che la temperatura di equilibrio tende a salire rispetto a quella iniziale. La velocità della termalizzazione è dovuta alla differenza tra la disposizione casuale iniziale e quella finale del materiale, infatti il gas è molto più simile a una disposizione casuale rispetto a un solido, che ha disposizione cristallina, o ad un liquido. Ma dopo un numeo sufficiente di passi anche il gas trova l'equilibrio attorno alla temperatura desiderata $T*=1.2$

```python
equilibration = np.loadtxt("equilibration_gas.dat",dtype=float)
nsteps = range(equilibration.size)

plt.plot(nsteps,equilibration)
plt.xlabel('Steps')
plt.ylabel('Temperature')
plt.title('Curva di equilibrazione della temperatura del gas')
plt.grid(True)
plt.show()
```
